```python
import sys; sys.path.insert(0, '..')
import pandas as pd
from config import settings

tracks = pd.read_parquet(settings.PROCESSED_DIR / 'tracks_features.parquet')
print(f'{len(tracks):,} tracks, {tracks.shape[1]} columns')
tracks.head()
```

```python
# Coverage: how many tracks charted on Billboard?
charted = tracks['billboard_peak_position'].notna()
print(f'Charted: {charted.sum():,} / {len(tracks):,} ({charted.mean():.1%})')
tracks['primary_genre'].value_counts()
```
