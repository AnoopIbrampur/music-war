Walk through the engineered features: era/tempo/duration buckets, audio

```python
import sys; sys.path.insert(0, '..')
import pandas as pd
from config import settings

tracks = pd.read_parquet(settings.PROCESSED_DIR / 'tracks_features.parquet')
tracks[['era', 'season_released', 'tempo_bucket', 'duration_bucket', 'audio_cluster',
        'composite_success_score']].describe(include='all')
```

```python
# What does each audio cluster sound like?
from src.processing.feature_engineer import AUDIO_FEATURES
profile = tracks.groupby('audio_cluster')[[f for f in AUDIO_FEATURES if f in tracks.columns]].mean()
profile.round(2)
```

```python
# Composite score vs raw popularity: the chart components shift the ranking
import plotly.express as px
px.scatter(tracks, x='spotify_popularity', y='composite_success_score',
           color=tracks['billboard_peak_position'].notna().map({True: 'charted', False: 'uncharted'}),
           title='Composite score vs Spotify popularity', template='plotly_dark',
           labels={'color': 'Billboard'})
```
